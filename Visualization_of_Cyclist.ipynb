Converting date columns to timedate objects.

```python
#Average and maximum casual ride lengths
df['ride_length'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
df.head()
```

### Ride Duration Analysis
 Investigate the average, median, or distribution of ride lengths (e.g.,ride_length). This could include comparisons by member_casual (member vs.casual riders) or week_day (weekday vs. weekend usage).

Calculating the average length of rides.

```python
# Basic Statistics
average_ride_length = df['ride_length'].mean()
median_ride_length = df['ride_length'].median()

# Grouped Analysis by Member Type
grouped = df.groupby('member_casual')['ride_length'].agg(['mean', 'median'])

# Display Results
print(f"Average Ride Length: {average_ride_length} minutes")
print(f"Median Ride Length: {median_ride_length} minutes")
print(grouped)
```

###Usage Patterns
Analyze how ride patterns vary by time (e.g., started_at, ended_at), day of the week (week_day), or rider type (member_casual). This can reveal peak usage times and days.

Analysing the distribution of rides across the days of the week, segmented by rider type to get insights into how riders use the bikes and for what purposes.

```python
#plotting a graph to show the distribution of rides across the days of the week
# Group the data by 'week_day' and 'member_casual', and count the rides
ride_counts = df.groupby(['week_day', 'member_casual']).size().unstack()

# Plotting
ride_counts.plot(kind='bar', stacked=False)

plt.title('Distribution of Rides Across Days of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)  # Rotates the x-axis labels to be more readable
plt.legend(title='Rider Type')
plt.show()
```

###Station Analysis

###Time Analysis
Understanding what times different clients prefer to use the bikes can be valuable in tailoring services and campaigns around peak times when they use the bike.

```python
# Extract the hour
df['hour'] = df['started_at'].dt.hour

# Group by hour and rider type, and count the rides
hourly_distribution = df.groupby(['hour', 'member_casual']).size().unstack()

# Plotting
hourly_distribution.plot(kind='bar', figsize=(12, 6))

plt.title('Popular Times of Day for Casual and Member Rides')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Rides')
plt.xticks(ticks=range(24), labels=[f"{h:02d}:00" for h in range(24)])
plt.legend(title='Rider Type')
plt.show()
```

```python
# Extract the hour
df['end_hour'] = df['ended_at'].dt.hour

# Group by end hour and rider type, and count the rides
end_hourly_distribution = df.groupby(['end_hour', 'member_casual']).size().unstack()

# Plotting
end_hourly_distribution.plot(kind='bar', figsize=(12, 6))

plt.title('Popular End Times of Day for Casual and Member Rides')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Rides')
plt.xticks(ticks=range(24), labels=[f"{h:02d}:00" for h in range(24)]) # Formatting hour labels
plt.legend(title='Rider Type')
plt.show()
```

##Time series Analysis
If your data spans a significant time period, you could look for trends or changes in usage over time.

```python
# Ensure that these DataFrames are treated as separate copies
casual_rides_df = casual_rides_df.copy()
member_rides_df = member_rides_df.copy()

# Extract month from 'started_at' for casual riders
casual_rides_df['month'] = casual_rides_df['started_at'].dt.to_period('M')

# Extract month from 'started_at' for member riders
member_rides_df['month'] = member_rides_df['started_at'].dt.to_period('M')

# Group by month and count the rides
casual_monthly_trends = casual_rides_df.groupby('month').size()
member_monthly_trends = member_rides_df.groupby('month').size()

# Plot the trends over time
plt.figure(figsize=(12, 6))

plt.plot(casual_monthly_trends.index.astype(str), casual_monthly_trends, marker='o', linestyle='-', label='Casual')
plt.plot(member_monthly_trends.index.astype(str), member_monthly_trends, marker='x', linestyle='--', label='Member')

plt.title('Monthly Usage Trends')
plt.xlabel('Month')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

###Comparative analysis of rideable type.

Compare different types of rideable (e.g., rideable_type) in terms of usage, distance, or duration.


```python
# Calculate the distance for each ride if not already present
if 'distance_km' not in df.columns:
    df['distance_km'] = df.apply(lambda row: haversine(row['start_lng'], row['start_lat'],
                                                        row['end_lng'], row['end_lat']), axis=1)

# Calculate usage by rideable type
usage_by_rideable = df['rideable_type'].value_counts()

# Calculate average distance by rideable type
average_distance_by_rideable = df.groupby('rideable_type')['distance_km'].mean()

# Assuming 'started_at' and 'ended_at' are already converted to datetime if they represent the start and end times
# Calculate duration for each ride if not already present
if 'duration' not in df.columns:
    df['duration'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60  # duration in minutes

# Calculate average duration by rideable type
average_duration_by_rideable = df.groupby('rideable_type')['duration'].mean()

# Combine the results into a single DataFrame
comparative_analysis = pd.DataFrame({
    'Usage': usage_by_rideable,
    'Average Distance (km)': average_distance_by_rideable,
    'Average Duration (minutes)': average_duration_by_rideable
})

# Reset index to make 'rideable_type' a column instead of an index
comparative_analysis.reset_index(inplace=True)
comparative_analysis.rename(columns={'index': 'Rideable Type'}, inplace=True)

# Display the results
print(comparative_analysis)
```
