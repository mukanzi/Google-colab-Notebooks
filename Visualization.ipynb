#Data visualizations for Cyclist

The code is to develop visualizations using Python, specifically Matplotlib to visualize relationships in the cyclist dataset.

## Importing and set up

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set()
```

## Setting up the dataframe

```python
df = pd.read_csv("/content/cyclist_cleaned_data.csv")
df.dtypes
```

```python
df['ride_length'] = df['ride_legth']
df.drop('ride_legth', axis=1, inplace=True)
df.head()
```

```python
df.drop("day_of_week", axis=1, inplace=True)
df.head()
```

```python
#dropping the first column
df['started_at']= pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
df.dtypes
```

```python
df.drop(df.columns[0], inplace=True,axis=1)
df.head()
```

```python
#Creating casual and member dfs for easier summary
casual_rides_df = df[df['member_casual']== 'casual']
member_rides_df = df[df['member_casual']=='member']
member_rides_df.head()
```

```python
#Average and maximum casual ride lengths
df['ride_length'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
df.head()
```

```python
# Basic Statistics
average_ride_length = df['ride_length'].mean()
median_ride_length = df['ride_length'].median()

# Grouped Analysis by Member Type
grouped = df.groupby('member_casual')['ride_length'].agg(['mean', 'median'])

# Display Results
print(f"Average Ride Length: {average_ride_length} minutes")
print(f"Median Ride Length: {median_ride_length} minutes")
print(grouped)
```

```python
#plotting a graph to show the distribution of rides across the days of the week
# Group the data by 'week_day' and 'member_casual', and count the rides
ride_counts = df.groupby(['week_day', 'member_casual']).size().unstack()

# Plotting
ride_counts.plot(kind='bar', stacked=False)

plt.title('Distribution of Rides Across Days of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)  # Rotates the x-axis labels to be more readable
plt.legend(title='Rider Type')
plt.show()
```

```python
# Group the data by 'rideable_type' and 'member_casual', and count the rides
ride_counts = df.groupby(['rideable_type', 'member_casual']).size().unstack()

# Plotting
ride_counts.plot(kind='bar', stacked=False)

plt.title('Rideable Type Distribution for Casual vs Member')
plt.xlabel('Rideable Type')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)  # Rotates the x-axis labels for better readability
plt.legend(title='Rider Type')
plt.show()
```

```python
# Calculate top 10 start and end stations for casual rides
top_casual_starts = casual_rides_df['start_station_name'].value_counts().head(10)
top_casual_ends = casual_rides_df['end_station_name'].value_counts().head(10)

# Calculate top 10 start and end stations for member rides
top_member_starts = member_rides_df['start_station_name'].value_counts().head(10)
top_member_ends = member_rides_df['end_station_name'].value_counts().head(10)

# Combine into a single DataFrame for display
top_stations = pd.DataFrame({
    'Casual Top Start Stations': top_casual_starts.index,
    'Casual Start Counts': top_casual_starts.values,
    'Casual Top End Stations': top_casual_ends.index,
    'Casual End Counts': top_casual_ends.values,
    'Member Top Start Stations': top_member_starts.index,
    'Member Start Counts': top_member_starts.values,
    'Member Top End Stations': top_member_ends.index,
    'Member End Counts': top_member_ends.values
})

# Display the table
print(top_stations)
```

```python
# Calculate top 10 end stations for casual rides
top_casual_ends = casual_rides_df['end_station_name'].value_counts().head(10)

# Calculate top 10 end stations for member rides
top_member_ends = member_rides_df['end_station_name'].value_counts().head(10)

# Combine into a single DataFrame for display
top_end_stations = pd.DataFrame({
    'Casual Top End Stations': top_casual_ends.index,
    'Casual End Counts': top_casual_ends.values,
    'Member Top End Stations': top_member_ends.index,
    'Member End Counts': top_member_ends.values
})

# Display the table
print(top_end_stations)
```

```python
# For casual_rides_df
# Create a route column
casual_rides_df['route'] = casual_rides_df['start_station_name'] + " to " + casual_rides_df['end_station_name']

# Count the occurrences of each route and get the top routes
top_casual_routes = casual_rides_df['route'].value_counts().head(10)

# For member_rides_df
# Create a route column
member_rides_df['route'] = member_rides_df['start_station_name'] + " to " + member_rides_df['end_station_name']

# Count the occurrences of each route and get the top routes
top_member_routes = member_rides_df['route'].value_counts().head(10)

# Display the top routes
print("Most Popular Routes for Casual Riders:")
print(top_casual_routes)
print("\nMost Popular Routes for Member Riders:")
print(top_member_routes)
```

```python
top_casual_routes.plot(kind='bar')
plt.title('Top 10 Routes for Casual Riders')
plt.xlabel('Routes')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45, ha='right')
plt.show()

# Plot for Member Riders
top_member_routes.plot(kind='bar', color='orange')
plt.title('Top 10 Routes for Member Riders')
plt.xlabel('Routes')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Extract the hour
df['hour'] = df['started_at'].dt.hour

# Group by hour and rider type, and count the rides
hourly_distribution = df.groupby(['hour', 'member_casual']).size().unstack()

# Plotting
hourly_distribution.plot(kind='bar', figsize=(12, 6))

plt.title('Popular Times of Day for Casual and Member Rides')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Rides')
plt.xticks(ticks=range(24), labels=[f"{h:02d}:00" for h in range(24)])
plt.legend(title='Rider Type')
plt.show()
```

```python
# Extract the hour
df['end_hour'] = df['ended_at'].dt.hour

# Group by end hour and rider type, and count the rides
end_hourly_distribution = df.groupby(['end_hour', 'member_casual']).size().unstack()

# Plotting
end_hourly_distribution.plot(kind='bar', figsize=(12, 6))

plt.title('Popular End Times of Day for Casual and Member Rides')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Rides')
plt.xticks(ticks=range(24), labels=[f"{h:02d}:00" for h in range(24)]) # Formatting hour labels
plt.legend(title='Rider Type')
plt.show()
```

```python
# Assuming you have a DataFrame df and have already extracted the hour as 'hour'
hourly_counts = df.groupby(['hour']).size()

# Convert hours to angles (in radians) for a polar plot
angles = np.linspace(0, 2 * np.pi, 24, endpoint=False).tolist()

# The plot will start at midnight and move clockwise
angles += angles[:1]

# Adding the count for 00:00 at the end to close the plot
hourly_data = hourly_counts.tolist() + hourly_counts[:1].tolist()

# Create a polar plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot the data
ax.fill(angles, hourly_data, color='blue', alpha=0.25)

# Set the direction of the zero angle
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Set up the grid and labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels([f"{h:02d}:00" for h in range(24)])

plt.title('Distribution of Rides Over a Day')
plt.show()
```

##Time series Analysis
If your data spans a significant time period, you could look for trends or changes in usage over time.

```python
# Ensure that these DataFrames are treated as separate copies
casual_rides_df = casual_rides_df.copy()
member_rides_df = member_rides_df.copy()

# Extract month from 'started_at' for casual riders
casual_rides_df['month'] = casual_rides_df['started_at'].dt.to_period('M')

# Extract month from 'started_at' for member riders
member_rides_df['month'] = member_rides_df['started_at'].dt.to_period('M')

# Group by month and count the rides
casual_monthly_trends = casual_rides_df.groupby('month').size()
member_monthly_trends = member_rides_df.groupby('month').size()

# Plot the trends over time
plt.figure(figsize=(12, 6))

plt.plot(casual_monthly_trends.index.astype(str), casual_monthly_trends, marker='o', linestyle='-', label='Casual')
plt.plot(member_monthly_trends.index.astype(str), member_monthly_trends, marker='x', linestyle='--', label='Member')

plt.title('Monthly Usage Trends')
plt.xlabel('Month')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
